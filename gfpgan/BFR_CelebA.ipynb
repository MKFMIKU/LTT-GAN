```python
from gfpgan.data.ffhq_degradation_dataset import FFHQDegradationDataset
from basicsr.utils import img2tensor, tensor2img
import cv2
import os
```

```python
import yaml

with open('/home/viu/Work/GFPGAN/options/ab_gfpgan_v1_simple.yml', 'r') as stream:
    try:
        parsed_yaml=yaml.safe_load(stream)
        data_config = parsed_yaml['datasets']['train']
    except yaml.YAMLError as exc:
        print(exc)
```

```python
data_config
```

```python
data_config['dataroot_gt'] = '/data/datasets/BFR/raw'
data_config['use_shuffle'] = False
data_config['use_hflip'] = False
# data_config['color_jitter_prob'] = 0
# data_config['color_jitter_pt_prob'] = 0
# data_config['gt_gray'] = True
```

```python
dataset = FFHQDegradationDataset(data_config)
```

```python
for batch in dataset:
    print(batch['gt_path'])
    lq = batch['lq']
    lq_img = tensor2img(lq.squeeze(0), rgb2bgr=True, min_max=(-1, 1))
    cv2.imwrite(os.path.join('/data/datasets/BFR/lq', os.path.basename(batch['gt_path'])), lq_img)
```
